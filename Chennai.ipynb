```python
print("Give the inputs of road segment(0<=limit<=240):")
print("Enter start point of road segment(0<=limit<=240):")
while(True):
    mini= int(input())
    if ((mini>240) or (mini<0)):
        print("Enter proper value")
        continue
    else:
        break
print("Enter end point of road segment(0<=limit<=240):")        
while(True):
    maxi= int(input())
    if ((maxi>240) or (maxi<0)):
        print("Enter proper value")
        continue
    else:
        break        
```

```python
print("Enter the point at which the information is required(mini<=limit<=maxi):")
while(True):
    inp= int(input())
    if ((inp<mini) or (inp>maxi)):
        print("Enter proper value (mini<=limit<=maxi)")
        continue
    else:
        break
```

```python
print("Enter the instant at which the information is required(1<=limit<=30):")
while(True):
    time= int(input())
    if ((time>31) or (time<1)):
        print("Enter proper value (1<=limit<=31)")
        continue
    elif((time<=15)):
        T_any_min=df.loc[df['Time (sec)'] <= time*60]
        break
    else:
        T_any_min=df1.loc[df1['Time (sec)'] <= (time-15)*60]
        break
```

```python
print("Density(veh/km):")
den*1000/(maxi-mini)
```

```python
Super_Final=[]
for i in range(1,61):
    if (i<31):
        T_Any_Min=df.loc[(df['Time (sec)'] <= i*60/2)&(df['Time (sec)'] >= (i-1)*60/2)]
        OO=T_Any_Min.loc[(T_Any_Min['Long Distance (m)'] >= mini)& (T_Any_Min['Long Distance (m)'] <= maxi)&(T_Any_Min['Time (sec)'] == i*60/2)]
        den=OO['Vehicle Number'].count()
    else:
        T_Any_Min=df1.loc[(df1['Time (sec)'] <= (i-30)*60/2)&(df1['Time (sec)'] >= (i-31)*60/2)]
        OO=T_Any_Min.loc[(T_Any_Min['Long Distance (m)'] >= mini)& (T_Any_Min['Long Distance (m)'] <= maxi)&(T_Any_Min['Time (sec)'] == (i-30)*60/2)]
        den=OO['Vehicle Number'].count()
        
    print(i)    
    k=len(T_Any_Min['Vehicle Number'].loc[(T_Any_Min['Long Distance (m)']<=maxi)& (T_Any_Min['Long Distance (m)']>=inp)].unique())
    s_1=T_Any_Min.loc[(T_Any_Min['Long Distance (m)'] > (inp-5))& (T_Any_Min['Long Distance (m)']<=(inp+5))]
    speed_1=[]
    for q in range(s_1['Vehicle Number'].min(),s_1['Vehicle Number'].max()+1):
        p=s_1[['Vehicle Number','Long Speed (m/sec)','Long Distance (m)']][T_Any_Min['Vehicle Number'] == q].mean(axis=0)
        speed_1.append(p)
    Speed_final_1=pd.DataFrame(speed_1, columns = ['Vehicle Number','Long Speed (m/sec)','Long Distance (m)'], index=None) 
    Speed_final_1.dropna(inplace=True)
    Speed_final_1['Hmean']=1/Speed_final_1['Long Speed (m/sec)']
    final=[Speed_final_1['Long Speed (m/sec)'].mean(axis=0),Speed_final_1['Hmean'].count()/Speed_final_1['Hmean'].sum(),den*1000/(maxi-mini),k*60/0.5,OO['Long Speed (m/sec)'].mean()]
    Super_Final.append(final)
```
